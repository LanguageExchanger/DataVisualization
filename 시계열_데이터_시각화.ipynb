```python
# Re-import necessary libraries after code execution state reset
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf

# Re-generate data
np.random.seed(42)
stationary = np.random.normal(loc=0, scale=1, size=200)
noise = np.random.normal(loc=0, scale=1, size=200)
non_stationary = np.cumsum(noise)
time = pd.date_range(start='2020-01-01', periods=200, freq='D')

df = pd.DataFrame({
    'Stationary': stationary,
    'Non-Stationary': non_stationary
}, index=time)

# Compute rolling mean and standard deviation
rolling_window = 20
df['Stationary_Mean'] = df['Stationary'].rolling(window=rolling_window).mean()
df['Stationary_Std'] = df['Stationary'].rolling(window=rolling_window).std()
df['NonStationary_Mean'] = df['Non-Stationary'].rolling(window=rolling_window).mean()
df['NonStationary_Std'] = df['Non-Stationary'].rolling(window=rolling_window).std()

# Plot rolling statistics
fig, axes = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

# Stationary time series with rolling stats
axes[0].plot(df.index, df['Stationary'], label='Stationary Series', color='blue')
axes[0].plot(df.index, df['Stationary_Mean'], label='Rolling Mean', color='green')
axes[0].plot(df.index, df['Stationary_Std'], label='Rolling Std', color='orange')
axes[0].set_title('✅ Stationary Series: Rolling Mean & Std Dev')
axes[0].legend()

# Non-stationary time series with rolling stats
axes[1].plot(df.index, df['Non-Stationary'], label='Non-Stationary Series', color='red')
axes[1].plot(df.index, df['NonStationary_Mean'], label='Rolling Mean', color='green')
axes[1].plot(df.index, df['NonStationary_Std'], label='Rolling Std', color='orange')
axes[1].set_title('❌ Non-Stationary Series: Rolling Mean & Std Dev')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import adfuller

# Seed for reproducibility
np.random.seed(42)

# Generate stationary series (white noise)
stationary = np.random.normal(loc=0, scale=1, size=200)

# Generate non-stationary series (random walk)
noise = np.random.normal(loc=0, scale=1, size=200)
non_stationary = np.cumsum(noise)

# Create time index
time = pd.date_range(start='2020-01-01', periods=200, freq='D')

# Create DataFrame
df = pd.DataFrame({
    'Stationary': stationary,
    'Non-Stationary': non_stationary
}, index=time)

# Plot
fig, axes = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

# Stationary plot
axes[0].plot(df.index, df['Stationary'], label='Stationary (White Noise)', color='blue')
axes[0].set_title('✅ Stationary Time Series')
axes[0].set_ylabel('Value')
axes[0].legend()

# Non-stationary plot
axes[1].plot(df.index, df['Non-Stationary'], label='Non-Stationary (Random Walk)', color='red')
axes[1].set_title('❌ Non-Stationary Time Series')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('Value')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Re-import necessary libraries after code execution state reset
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf

# Re-generate data
np.random.seed(42)
stationary = np.random.normal(loc=0, scale=1, size=200)
noise = np.random.normal(loc=0, scale=1, size=200)
non_stationary = np.cumsum(noise)
time = pd.date_range(start='2020-01-01', periods=200, freq='D')

df = pd.DataFrame({
    'Stationary': stationary,
    'Non-Stationary': non_stationary
}, index=time)

# Plot ACF for both time series
fig, axes = plt.subplots(2, 1, figsize=(12, 8))

# ACF of stationary series
plot_acf(df['Stationary'], lags=40, ax=axes[0])
axes[0].set_title('ACF of ✅ Stationary Time Series')

# ACF of non-stationary series
plot_acf(df['Non-Stationary'], lags=40, ax=axes[1])
axes[1].set_title('ACF of ❌ Non-Stationary Time Series')

plt.tight_layout()
plt.show()
```
